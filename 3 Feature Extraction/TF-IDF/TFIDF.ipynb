### Testing & Vizualization

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

def test_feature_extraction_TFIDF():
    # Define your corpus
    corpus = [
        'large bbq pulled pork', 
        'large pie with green pepper and with extra peperonni', 
        'large pie green pepper extra peperonni', 
        'party size stuffed crust pie american cheese mushroom', 
        'one personal sized artichoke'
    ]
    
    # Instantiate feature extractor
    TFIDF_extractor = TfidfExtractor(corpus)

    # Test feature extraction
    feature_names, representation = TFIDF_extractor.extract()
    
    # Create a DataFrame to visualize the TF-IDF values
    tfidf_df = pd.DataFrame(representation, columns=feature_names, index=[f"D{i+1}" for i in range(len(corpus))])
    
    # Adjust figure size and font size to fit everything properly
    plt.figure(figsize=(12, 8))  # Increase the figure size for better spacing
    sns.heatmap(tfidf_df, annot=True, fmt=".2f", cmap="Purples", cbar=False, linewidths=0.5, annot_kws={"size": 10})
    plt.title('TF-IDF Representation', fontsize=16)
    plt.xlabel('Words', fontsize=12)
    plt.ylabel('Documents', fontsize=12)
    plt.xticks(rotation=90, fontsize=10)  # Rotate x-axis labels for better readability
    plt.yticks(rotation=0, fontsize=10)  # Set y-axis label font size and no rotation
    plt.tight_layout()  # Adjust layout to prevent label cutoff
    plt.show()

# Run the test with visualization
test_feature_extraction_TFIDF()
```
