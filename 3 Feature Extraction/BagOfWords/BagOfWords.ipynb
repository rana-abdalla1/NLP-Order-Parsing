# Feature Extraction: Bag of Words

### Imports

```python
from BagOfWords import BagOfWordsExtractor
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Testing & Vizualization

```python
def test_feature_extraction_BagOfWords():
    # Define your corpus
    corpus = [
        'large bbq pulled pork', 
        'large pie with green pepper and with extra peperonni', 
        'large pie green pepper extra peperonni', 
        'party size stuffed crust pie american cheese mushroom', 
        'one personal sized artichoke'
    ]
    
    # Instantiate feature extractor
    bow_extractor = BagOfWordsExtractor(corpus)

    # Test feature extraction
    feature_names, representation = bow_extractor.extract()
    
    # Create a DataFrame for better visualization
    df = pd.DataFrame(representation, columns=feature_names, index=[f"D{i+1}" for i in range(len(corpus))])
    
    # Visualize the representation as a heatmap
    plt.figure(figsize=(12, 6))
    sns.heatmap(df, annot=True, fmt="d", cmap="Purples", cbar=False)
    plt.title("Bag of Words Representation")
    plt.xlabel("Words")
    plt.ylabel("Documents")
    plt.xticks(rotation=90)
    plt.show()

# Run the test for feature extraction with visualization
test_feature_extraction_BagOfWords()
```
